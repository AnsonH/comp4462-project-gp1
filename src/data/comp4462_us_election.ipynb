# Import Dataset

```python
import pandas as pd
import numpy as np
import json

data = pd.read_csv("1976-2020-president.csv",encoding="ISO-8859-1")
data
```

# Data Analysis

```python
states = data.groupby('state')
states = list(states.groups)

year = 1976
results = []

while year <= 2020:
    temp = []
    for s in states:
        yearly_total_democrat = data.loc[((data['party_detailed'] == "DEMOCRAT") & (data['year'] == year) & (data['state'] == s)), "candidatevotes"].sum() 
        yearly_total_republican = data.loc[((data['party_detailed'] == "REPUBLICAN") & (data['year'] == year) & (data['state'] == s)), "candidatevotes"].sum()

        winner = "DEMOCRAT" if yearly_total_democrat > yearly_total_republican else "REPUBLICAN"
        temp.append({
            "state": "WASHINGTON, D.C." if s == "DISTRICT OF COLUMBIA" else str(s),
            "winner": winner, 
            "democrat_total": int(yearly_total_democrat),
            "republican_total": int(yearly_total_republican)    
        })
        
    results.append({
        "year": year,
        "details": temp
    })

    year += 4

data.groupby(['party_detailed'])

with open('us_election.json', 'w') as json_file:
    json.dump(results, json_file, indent=4)
```
